```python
import matplotlib.pyplot as plt # type: ignore

# ข้อมูล x และ y
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

# สร้างกราฟ
plt.plot(x, y, marker='o', linestyle='-', color='dodgerblue', markersize=8, linewidth=2, label='Data Line')

# เพิ่มชื่อแกน
plt.xlabel('X Axis', fontsize=14, fontweight='bold')
plt.ylabel('Y Axis', fontsize=14, fontweight='bold')
plt.title('Simple Line Plot', fontsize=16, fontweight='bold')

# แสดงคำอธิบายกราฟ
plt.legend(fontsize=10)

# ปรับขนาดของตัวเลขที่แกน
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# เพิ่มกรอบให้กับกราฟ
plt.gca().spines['top'].set_visible(False)  # ซ่อนขอบด้านบน
plt.gca().spines['right'].set_visible(False)  # ซ่อนขอบด้านขวา

# เพิ่มกรอบเงา
plt.gca().patch.set_facecolor('lightgray')

# แสดงผล
plt.show()
```

```python
import matplotlib.pyplot as plt # type: ignore

x = [1, 2, 3, 4, 5]
y = [5, 7, 9, 11, 13]

plt.step(x, y, where='mid', color='purple')
plt.title('Step Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt # type: ignore

# ข้อมูล
categories = ['red', 'green', 'blue', 'purple']
values = [100, 200, 300, 400]

# สร้างกราฟแท่ง
bars = plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'], edgecolor='black', linewidth=1.5, alpha=0.8)

# เพิ่มการสร้างเงาด้วยการใช้ alpha และการเลื่อนตำแหน่ง (การทำให้กราฟมีมิติ)
for bar in bars:
    bar.set_alpha(0.7)  # ทำให้โปร่งใสเล็กน้อย

# ปรับรูปแบบชื่อแกนและชื่อกราฟ
plt.xlabel('Categories', fontsize=14, fontweight='bold')
plt.ylabel('Values', fontsize=14, fontweight='bold')
plt.title('TEST Bar Graph', fontsize=16, fontweight='bold')

# ปรับขนาดและสไตล์ของตัวเลขบนแกน y
plt.yticks(fontsize=12)

# เพิ่มกรอบให้กับกราฟ
plt.gca().spines['top'].set_visible(False)  # ซ่อนขอบด้านบน
plt.gca().spines['right'].set_visible(False)  # ซ่อนขอบด้านขวา

# แสดงผล
plt.show()
```

```python
import matplotlib.pyplot as plt # type: ignore

# ข้อมูล
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 40, 20, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# เพิ่มการเน้นให้กับชิ้นส่วนของกราฟ (explode)
explode = (0.1, 0, 0, 0)  # เน้น Apple โดยการขยายออกมาเล็กน้อย

# สร้างกราฟวงกลม
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})

# ตั้งชื่อกราฟ
plt.title('Simple Pie Chart', fontsize=16, fontweight='bold')

# แสดงผล
plt.show()
```
