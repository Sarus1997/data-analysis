```python
import matplotlib.pyplot as plt

# ข้อมูล x และ y
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

# สร้างกราฟ
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Data Line')

# เพิ่มชื่อแกน
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Plot')

# แสดงคำอธิบายกราฟ
plt.legend()

# แสดงผล
plt.show()
```

```python
import matplotlib.pyplot as plt

# ข้อมูล
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# สร้างกราฟแท่ง
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'])

# เพิ่มชื่อแกนและชื่อกราฟ
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')

# แสดงผล
plt.show()
```

```python
import matplotlib.pyplot as plt

# ข้อมูล
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 40, 20, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# สร้างกราฟวงกลม
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# แสดงผล
plt.title('Simple Pie Chart')
plt.show()
```
